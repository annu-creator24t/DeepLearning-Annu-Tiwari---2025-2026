```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

# Features
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Target
y = pd.Series(iris.target, name="target")

# Combine
df = pd.concat([X, y], axis=1)

df.head()
```
